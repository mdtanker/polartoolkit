# Create a simple map

Here we will showcase creating a very simple map with some gridded data. All of the plotting functions in PolarToolkit use the python package PyGMT, which is a python wrapper on the wide-used GMT (Generic-Mapping-Tools).

Import the modules

```python
from polartoolkit import fetch, maps
```

Use the PolarToolkit `fetch` module to download some Antarctic geothermal heat flux data and return the grid as an xarray.DataArray

```python
grid = fetch.ghf(
    version="burton-johnson-2020",
)
```

Create a simple map with default settings

```python
fig = maps.plot_grd(grid)

fig.show(dpi=200)
```
