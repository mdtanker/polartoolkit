**Associated paper:**

```python
%%capture
%load_ext autoreload
%autoreload 2

import polartoolkit as ptk
```

```python
data = ptk.fetch.deepbedmap(
    # available options
    region=ptk.regions.minna_bluff,
    # registration ("g" for gridline or "p" for pixel),
)
data
```

```python
_ = ptk.get_grid_info(data, print_info=True)
```

```python
fig = ptk.plot_grid(
    data,
    title="DeepBedMap Bed Topography",
    coast=True,
    cbar_label="meters",
    hist=True,
    robust=True,
    cmap="rain",
    reverse_cpt=True,
    hemisphere="south",
)

fig.show(dpi=200)
```
