```python
# instead of passing the imagery filepath to `plot_grid`, we can use the keyword argument
# `imagery_basemap=True` to either `maps.plot_grd()` or `maps.basemap()`
fig = maps.basemap(
    imagery_basemap=True,
    title="LIMA imagery",
    coast=True,
    coast_pen="0.6p,salmon",
    colorbar=False,
    hemisphere="south",
)
fig.show(dpi=200)
```
