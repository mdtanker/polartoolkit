```python
from __future__ import annotations

%%capture
%load_ext autoreload
%autoreload 2

from polartoolkit import fetch, maps, utils
```

```python
data = fetch.gia(
    version="stal-2020",
    # available options
    # region,
    # spacing,
    # registration ("g" for gridline or "p" for pixel),
)
data
```

```python
_ = utils.get_grid_info(data, print_info=True)
```

```python
fig = maps.plot_grd(
    data,
    title="Glacial Isostatic Adjustment",
    coast=True,
    cbar_label="meters",
    hist=True,
    robust=True,
    hemisphere="south",
)

fig.show(dpi=200)
```
