```python
%%capture
%load_ext autoreload
%autoreload 2

import os

import polartoolkit as ptk
```

```python
# set default to northern hemisphere for this notebook
os.environ["POLARTOOLKIT_HEMISPHERE"] = "north"
```

```python
fpath = ptk.fetch.modis(
    # choose which resolution to use
    version="500m",
    # version="100m"
)
fpath
```

```python
# instead of passing the imagery filepath to `plot_grid`, we can use the keyword argument
# `modis_basemap=True` to either `plot_grid()` or `basemap()`
fig = ptk.basemap(
    region=ptk.regions.greenland,
    modis_basemap=True,
    title="MODIS Mosaic of Greenland",
    modis=True,  # this automatically uses a nice colorscale for modis data
)
fig.show(dpi=200)
```
